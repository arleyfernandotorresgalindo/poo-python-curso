## Método `__str__` y `__repr__`

````python
class Persona1: 

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad 

    def __repr__(self):
        return f'Esta persona se llama: {self.nombre}'     

    def __str__(self): 
        return f'({self.nombre}, {self.edad})' 
    
````

````python
P1, P2, P3 = Persona1('Ana', 35), Persona1('Carlos', 33), Persona1('Maria', 41)
````

## Vamos a crear la clase de vectores en $R^2$: `__add__` y `__sub__` 

````python
class VectoresR2:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def SumarVectores(self, v1):
        x1 = self.x + v1.x
        y1 = self.y + v1.y
        return VectoresR2(x1, y1)
    
    def __add__(self, v1):
        x1 = self.x + v1.x
        y1 = self.y + v1.y
        return VectoresR2(x1, y1)

    def RestarVectores(self, v1):
        x1 = self.x - v1.x
        y1 = self.y - v1.y
        return VectoresR2(x1, y1)
    
    def __sub__(self, v1):
        x1 = self.x - v1.x
        y1 = self.y - v1.y
        return VectoresR2(x1, y1)
    
    def MultiplicacionEscalar(self, escalar):
        x1 = escalar*self.x
        y1 = escalar*self.y
        return VectoresR2(x1, y1)  

    def __rmul__(self, escalar):
        x1 = escalar*self.x
        y1 = escalar*self.y
        return VectoresR2(x1, y1)  
    
    def Producto_Punto(self, v1):
        x1 = self.x*v1.x
        y1 = self.y*v1.y
        return x1 + y1

    def __mul__(self, v1):
        x1 = self.x*v1.x
        y1 = self.y*v1.y
        return x1 + y1




    def __repr__(self):
        return f'V2({self.x}, {self.y})'
````

````python
V1, V2 = VectoresR2(2,3), VectoresR2(5,7)
V1.SumarVectores(V2), V1.RestarVectores(V2)
````

````python
V1 + V2, V1 - V2, V1.MultiplicacionEscalar(2), 2*V1, V1.Producto_Punto(V2), V1*V2
````
