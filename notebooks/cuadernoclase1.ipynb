# ⚙️ POO II: Métodos de instancia y comportamiento

Ahora introduces comportamiento:

Los objetos no solo tienen datos, también hacen cosas.

```python
class Estudiante1:
    def __init__(self, nombre, edad, promedio):
        self.nombre = nombre
        self.edad = edad
        self.promedio = promedio
    
    def saludar(self):
        print(f"Hola, soy {self.nombre}")
    
    def aprobar(self):
        return self.promedio >= 3.0

````python
class Estudiante1:
    def __init__(self, nombre, edad, promedio):
        self.nombre = nombre
        self.edad = edad
        self.promedio = promedio
    
    def saludar(self):
        print(f"Hola, soy {self.nombre}")
    
    def aprobar(self):
        return self.promedio >= 3.0
````

````python
e1 = Estudiante1("Ana", 20, 4.5)

e1.saludar()        
print(e1.aprobar()) 
````

````python
vars(e1)
````

````python
metodos = [m for m in dir(e1) 
           if callable(getattr(e1, m)) and not m.startswith("__")]

print(metodos)
````

# 🧠 POO III: Métodos especiales y protocolos básicos

En Python, las clases pueden definir **métodos especiales** (*dunder methods*, porque usan `__`).

Estos métodos permiten que nuestros objetos se comporten como tipos nativos.

---

# 🔹 1. Método `__init__`

```python

    def __init__(self, nombre, edad):

        self.edad = edad
```

👉 Se ejecuta automáticamente al crear el objeto.

## Método `__str__` y `__repr__`

````python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre
    
    def __str__(self):
        return f"Persona: {self.nombre}"

p = Persona("Ana")
print(p)
````
